```python
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def Plot_2images(imagen01,imagenResultante,imagen02):
  plt.subplot(131)
  plt.imshow(cv2.cvtColor(imagen01, cv2.COLOR_BGR2RGB))
  plt.title('Imagen 1')
  plt.xticks([])
  plt.yticks([])

  plt.subplot(132)
  plt.imshow(cv2.cvtColor(imagenResultante, cv2.COLOR_BGR2RGB))
  plt.title('Imagen Resultante')
  plt.xticks([])
  plt.yticks([])

  plt.subplot(133)
  plt.imshow(cv2.cvtColor(imagen02, cv2.COLOR_BGR2RGB))
  plt.title('Imagen 2')
  plt.xticks([])
  plt.yticks([])

  plt.show()
```

```python
def Plot_1image(imagen01,imagenResultante):
  plt.subplot(131)
  plt.imshow(cv2.cvtColor(imagen01, cv2.COLOR_BGR2RGB))
  plt.title('Imagen 1')
  plt.xticks([])
  plt.yticks([])

  plt.subplot(132)
  plt.imshow(cv2.cvtColor(imagenResultante, cv2.COLOR_BGR2RGB),vmax=255)
  plt.title('Imagen Resultante')
  plt.xticks([])
  plt.yticks([])

  plt.show()
```

```python
def switch():
  cont = 1  

  for i in range (20):
    letra_E=input()
    if letra_E == 'e':
       
      '''1) METODO DE SUMA SIMPLE'''
      def suma_simple():
        #Operacion
        imaResult=cv2.add(imagen1,ima2)
        #Guarda imagen
        #cv2.imwrite('suma_simple.png',imaResult)
        #Ploteo de imagenes
        print("Suma Simple")
        Plot_2images(imagen1,imaResult,ima2) 

      '''2) METODO DE SUMA POR PESOS'''
      def suma_peso()  :    
        #Operaciones
        imaResult=cv2.addWeighted(imagen1,0.4,ima2,0.6,0)
        #Guarda imagien
        #cv2.imwrite('suma_peso.png',imaResult)  
        #Ploteo de imagenes
        print("Suma por Pesos")
        Plot_2images(imagen1,imaResult,ima2)

      '''3) METODO DE RESTA SIMPLE'''
      def resta_simple():
        #Operaciones
        imaResult=cv2.subtract(imagen1,ima2)
        #Guarda imagen
        #cv2.imwrite('resta_simple.png',imaResult)
        #Ploteo de imagenes
        print("Resta Simple")
        Plot_2images(imagen1,imaResult,ima2)

      '''4) METODO DE RESTA ABSOLUTA'''
      def resta_abs():
        #Operaciones
        imaResult=cv2.absdiff(imagen1,ima2)
        #Guarda imagen
        #cv2.imwrite('resta_abs.png',imaResult)  
        #Ploteo de imagenes
        print("Resta Absoluta")
        Plot_2images(imagen1,imaResult,ima2)

      '''5) METODO DE DIVISION'''
      def division():
        #Operaciones
        imaResult=cv2.divide(imagen1,ima2)
        #Guarda imagen
        #cv2.imwrite('division.png',imaResult)  
        #Ploteo de imagenes
        print("Division")
        Plot_2images(imagen1,imaResult,ima2)

      '''6) METODO DE MULTIPLICACION'''
      def multiplica():
        #Operaciones
        imaResult=cv2.multiply(imagen1,ima2)
        #Guarda imagen
        #cv2.imwrite('multiplica.png',imaResult)  
        #Ploteo de imagenes
        print("Multiplicacion")
        Plot_2images(imagen1,imaResult,ima2)

      '''7) METODO DE LOGARITMICA'''
      def logaritmo():
        #Operaciones
        logaritmo = np.zeros(imagen1.shape, imagen1.dtype)
        c = 1
        logaritmo = c * np.log(1 + imagen1)
        maxi = np.amax(logaritmo)
        imaResult=np.uint8(logaritmo / maxi * 255)                
        #Guarda imagen 
        #cv2.imwrite('logaritmo.png',imaResult)  
        #Ploteo de imagenes
        print("Logaritmo")
        Plot_1image(imagen1,imaResult)

      '''8) METODO DE RAIZ CUADRADA'''
      def raiz():
        #Operaciones
        imaResult= (imagen1**(0.5))  
        #Resize
        width = int(imagen1.shape[1])
        height = int(imagen1.shape[0])
        dsize=(width, height)
        imaResult = cv2.resize(imagen2, dsize)             
        #Guarda imagen
        #cv2.imwrite('raiz.png',imaResult)  
        #Ploteo de imagenes
        print("Raiz Cuadrada")
        Plot_1image(imagen1,imaResult)

      '''9) METODO DE DERIAVADA LAPLACIANA'''
      def deriv_lap():
        #Operaciones
        imaResult=cv2.Laplacian(imagen1,cv2.CV_32F)
        #Guarda imagenes
        #cv2.imwrite('deriv_lap.png',imaResult)  
        #Ploteo de imagenes
        print("Derivada Laplaciana")
        Plot_1image(imagen1,imaResult)           

      '''10) METODO DE DERIVADA SOBEL EN X'''
      def deriv_sobelX():
        #Operaciones
        imaResult=cv2.Sobel(imagen1,cv2.CV_32F,1,0,ksize=3)
        #Guarda imagen
        #cv2.imwrite('deriv_sobelX.png',imaResult)  
        #Ploteo de imagenes
        print("Derivada Sobel X")
        Plot_1image(imagen1,imaResult)       

      '''11) METODO DE DERIVADA SOBEL EN Y'''
      def deriv_sobelY():
        #Operaciones
        imaResult=cv2.Sobel(imagen1,cv2.CV_32F,0,1,ksize=3)
        #Guarda imagen
        #cv2.imwrite('derv_sobelY.png',imaResult)  
        #Ploteo de imagenes
        print("Derivada Sobel Y")
        Plot_1image(imagen1,imaResult)            

      '''12) METODO DE POTENCIA/EXPONENCIAL'''
      def potencia():
        #Operaciones
        imaResult=cv2.pow(imagen1,3)
        #Guarda imagen
        #cv2.imwrite('potencia.png',imaResult) 
        #Ploteo de imagenes
        print("Potencia")
        Plot_1image(imagen1,imaResult)

      '''13) METODO DE CONJUNCION'''
      def conjuncion():
        #Operaciones
        imaResult=cv2.bitwise_and(imagen1,ima2)
        #Guarda imagen
        #cv2.imwrite('conjuncion.png',imaResult)  
        #Ploteo de imagenes
        print("Conjuncion")
        Plot_2images(imagen1,imaResult,ima2)

      '''14) METODO DE DISYUNCION'''
      def disyuncion():
        #Operaciones
        imaResult=cv2.bitwise_or(imagen1,ima2)
        #Guarda imagen
        #cv2.imwrite('disyuncion.png',imaResult)  
        #Ploteo de imagenes
        print("Disyuncion")
        Plot_2images(imagen1,imaResult,ima2)

      '''15) METODO DE NEGACION'''
      def negacion():
        #Operaciones
        imaResult=cv2.bitwise_not(imagen1,)
        #Guarda imagen
        #cv2.imwrite('negacion.png',imaResult)  
        #Ploteo de imagenes
        print("Negacion")
        Plot_1image(imagen1,imaResult)

      '''16) METODO DE TRASLACION'''
      def tras_simple():
        #Operaciones
        ancho = imagen1.shape[1] #columnas
        alto = imagen1.shape[0] # filas
        #Traslacion
        M = np.float32([[1,0,60],[0,1,70]])
        imaResult = cv2.warpAffine(imagen1,M,(ancho,alto))
        #Guarda imagen
        #cv2.imwrite('tras_simple.png',imaResult)  
        #Ploteo de imagenes
        print("Traslacion")
        Plot_1image(imagen1,imaResult)

      '''17) METODO DE ESCALADO'''
      def escalado():
        #operaciones
        imaResult=cv2.resize(imagen1,(600,300), interpolation=cv2.INTER_CUBIC)
        #Guarda imagen
        #cv2.imwrite('escalado.png',imaResult)  
        #Ploteo de imagenes
        print("Escalado")
        Plot_1image(imagen1,imaResult)

      '''18) METODO DE ROTACION'''
      def rotacion():
        #Operaciones
        ancho = imagen1.shape[1] #columnas
        alto = imagen1.shape[0] # filas
        #Rotacion
        M = cv2.getRotationMatrix2D((ancho//2,alto//2),15,1)  #a 15° de rot
        imaResult = cv2.warpAffine(imagen1,M,(ancho,alto))
        #Guarda imagen
        #cv2.imwrite('rotacion.png',imaResult)    
        #Ploteo de imagenes
        print("Rotacion")
        Plot_1image(imagen1,imaResult)

      '''19) METODO DE TRASLACION A FIN'''
      def tras_afin():
        #Operaciones
        ancho = imagen1.shape[1] #columnas
        alto = imagen1.shape[0] # filas
        #Traslacion
        M = np.float32([[1,0,100],[0,1,150]])
        imaResult = cv2.warpAffine(imagen1,M,(ancho,alto))
        #Guarda imagen
        #cv2.imwrite('tras_afin.png',imaResult)  
        #Ploteo de imagenes
        print("Traslacion a fin")
        Plot_1image(imagen1,imaResult)

      '''20) METODO DE TRANSPUESTA'''
      def transpuesta():
        #operaciones
        imaResult=cv2.transpose(imagen1)
        #Guarda imagen
        #cv2.imwrite('transpuesta.png',imaResult)  
        #Ploteo de imagenes
        print("Transpuesta")
        Plot_1image(imagen1,imaResult)

      '''21) METODO DE PROYECCION'''
      def proyeccion():
        #Operaciones
        rows1, cols1, ch = imagen1.shape # Filas, Columnas, Canal = Tamaño de la imagen
        # Esquinas: Sup. Izq; Sup. Der; Inf. Izq; Inf. Der
        pts1 = np.float32([[0, 0], [300, 0], [0, 300],  [300, 300]])
        pts2 = np.float32([[70, 70], [230, 70], [0, 300], [300, 300]])
        M = cv2.getPerspectiveTransform(pts1, pts2)
        persp = cv2.warpPerspective(imagen1, M, (cols1, rows1))
        imaResult=persp + imagen1
        #Guarda imagen
        #cv2.imwrite('proyeccion.png',imaResult)  
        #Ploteo de imagenes
        print("Proyeccion")
        Plot_1image(imagen1,imaResult)

      def default():
        print("Opcion incorrecta")  
      
      dict={
          1:suma_simple,
          2:suma_peso,
          3:resta_simple,
          4:resta_abs,
          5:division,
          6:multiplica,
          7:logaritmo,
          8:raiz,
          9:deriv_lap,
          10:deriv_sobelX,
          11:deriv_sobelY,
          12:potencia,
          13:conjuncion,
          14:disyuncion,
          15:negacion,
          16:tras_simple,
          17:escalado,
          18:rotacion,
          19:tras_afin,
          20:transpuesta,
          21:proyeccion   
      }

      dict.get(cont,default)()
      cont = cont+1


#Imagenes para operaciones
imagen1=cv2.imread('/content/mascotas-perro-gatojpg.jpg')
imagen2=cv2.imread('/content/perro-salchicha.jpg')

#Modifica el tamaño de la seguna imagen para tener las mismas dimensiones
width = int(imagen2.shape[1]-86)
height = int(imagen2.shape[0]-12)
dsize = (width, height)
#Funcion para redimensionar
ima2 = cv2.resize(imagen2, dsize)
#se manda llamar la funcion switch
switch()      
```
